```python
from pathlib import Path
import urllib.request
import tarfile
import zipfile
import pandas as pd
```

```python
#LOAD DỮ LIỆU TỪ FILE ZIP
def load_data(zip_path, file_in_zip):
    with zipfile.ZipFile(Path(zip_path)) as z:
        with z.open(file_in_zip) as f:
            return pd.read_csv(f)
db=load_data("../../data/raw/data_kaggle_date.zip","diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
```
