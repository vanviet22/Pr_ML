```python
#load data 02
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv("../../data/processed/DB02.csv")
```

```python
#chuẩn bị data để huân luyện 
from sklearn.model_selection import train_test_split
x=df.drop('Diabetes',axis=1)
y=df['Diabetes']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.head()
```

```python
y.head()
```

```python
print(x.shape)
y.shape
```

```python
cols_to_scale = ['BMI', 'GenHlth', 'MentHlth','Age','Education','Income']
```

```python
#Scale dữ liệu
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[cols_to_scale]=sc.fit_transform(x_train[cols_to_scale])
x_train
```

```python
x_test[cols_to_scale]=sc.transform(x_test[cols_to_scale])
x_test
```

```python
#đánh giá training bới cross-validation với k-fold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,f1_score,classification_report, confusion_matrix, accuracy_score
# Hàm đánh giá mô hình với các chỉ số chung
def evaluate_model(model, x_train, y_train, x_test, y_test):
    # Cross-validation với k=5 và tiêu chí đánh giá là 'accuracy'
    cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
    print(f"{model.__class__.__name__} - Cross-validation scores:", cv_scores)
    print(f"Mean cross-validation score: {cv_scores.mean()}\n")
    
    # Dự đoán trên tập kiểm tra
    y_pred = model.predict(x_test)
    
    # Các chỉ số đánh giá trên tập kiểm tra
    print(f"{model.__class__.__name__} - Accuracy Score:", accuracy_score(y_test, y_pred))
    print(f"{model.__class__.__name__} - F1 Score:", f1_score(y_test, y_pred, average='weighted'))
    print(f"{model.__class__.__name__} - Classification Report:\n", classification_report(y_test, y_pred))
    print(f"{model.__class__.__name__} - Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("-" * 50)
```

```python
#thuật toán RandomForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
param_grid = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False] #lấy mẫu lại
}
rf=RandomForestClassifier()
rf_ramdom=RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=100,random_state=42, n_jobs=-1)
```

```python
rf_ramdom.fit(x_train,y_train)
```

```python
yRF_predict=rf_ramdom.predict(x_test)
```

```python
#print best hyperparameter
rf_ramdom.best_params_
```

```python
evaluate_model(rf_ramdom, x_train, y_train, x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'penalty': ['l1', 'l2'],#loại chuẩn hóa để tránh overfitting
    'C': [0.001, 0.01, 0.1, 1, 10, 100],#Điều khiển độ ngược chuẩn hóa
    'solver': ['liblinear'],#thuật toán tối ưu hóa hàm mất mát
    'class_weight': [None, 'balanced'],#giải quyết vấn đề dữ liệu mất cân bằng
    'max_iter': [1000, 1500, 2000]
}
lg=LogisticRegression()
lg_grid=GridSearchCV(estimator=lg, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
lg_grid.fit(x_train,y_train)
```

```python
lg_grid.predict(x_test)
```

```python
lg_grid.best_params_
```

```python
evaluate_model(lg_grid, x_train, y_train, x_test, y_test)
```

```python
#Sử dụng Ensemble learing Stacking
from sklearn.ensemble import StackingClassifier
# Khởi tạo StackingClassifier
stacking_clf = StackingClassifier(
    estimators=[('rf', rf_ramdom), ('lg', lg_grid)],
    final_estimator=LogisticRegression(max_iter=1000,random_state=42)
)
```

```python
stacking_clf.fit(x_train, y_train)
```

```python
yStack_pred=stacking_clf.predict(x_test)
```

```python
# Đánh giá mô hình Stacking
evaluate_model(stacking_clf, x_train, y_train, x_test, y_test)
```

```python
#fine tune final estimate of stacking 
from sklearn.ensemble import StackingClassifier
# Khởi tạo StackingClassifier
stacking_clf_finetune = StackingClassifier(
    estimators=[('rf', rf_ramdom), ('lg', lg_grid)],
    final_estimator=lg_grid
)
```

```python
stacking_clf_finetune.fit(x_train, y_train)
```

```python
yStack_finetune_pred=stacking_clf.predict(x_test)
```

```python
# Đánh giá mô hình Stacking với SVC
evaluate_model(stacking_clf_finetune, x_train, y_train, x_test, y_test)
```

```python
#fine tune final estimate of stacking is rf_random
from sklearn.ensemble import StackingClassifier
# Khởi tạo StackingClassifier
stacking_clf_finetuneRF = StackingClassifier(
    estimators=[('rf', rf_ramdom), ('lg', lg_grid)],
    final_estimator=rf_ramdom
)
```

```python
stacking_clf_finetuneRF.fit(x_train, y_train)
```

```python
# Đánh giá mô hình Stacking với SVC
evaluate_model(stacking_clf_finetuneRF, x_train, y_train, x_test, y_test)
```
